Code cells save the bronze data as 1000 CSV files to Azure Data Lake, and display the saved files for verification.

```python
%run "./storage_key"
```

```python
from pyspark.sql.functions import lit

storage_account_name = "konstyantyn"
container_name = "konstyantyn"
```

```python

# Output path
output_path = f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net/raw_lab3_1000_csv/" 

# Loading bronze data
df_1000 = spark.read.table("dbr_dev.konstyantyn_bronze.bronze_data")

# Separation of the files into 1000 CSVs
(df_1000.repartition(1000)
    .write\
    .mode("overwrite")\
    .option("header", "true")\
    .csv(output_path, mode='overwrite')
)

print("Data saved to: ", output_path)
```

The following code creates a new CSV file with an additional test column for streaming validation.

```python
# creating new file for streaming to check if it works

df_1001 = spark.read.table("dbr_dev.konstyantyn_bronze.bronze_data").limit(1)
new_file_df = df_1001.withColumn("new_test_column", lit("LAB3_TEST"))

# Define new file name for test output
new_test_file_path = f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net/raw_lab3_1000_csv/new_test_file.csv"

(new_file_df
    .write
    .mode("overwrite")
    .option("header", "true")
    .csv(new_test_file_path)
)

print("New test file saved to: ", new_test_file_path)
```
